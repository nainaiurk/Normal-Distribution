```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.ylim(0,1.2)
plt.title('Generalized Form of Normal Distribution')
plt.xlabel('x')
plt.ylabel('Φ(x,μ,σ)')
#####
m=0
s=0.5
x = np.linspace(-10,10,1000)
y = 1/(s*np.sqrt(2*np.pi))*(np.exp(-((x-m)**2)/(2*(s**2))))
plt.plot(x,y,label='μ=0,σ=0.5')
#####
m=0
s=1
y = 1/(s*np.sqrt(2*np.pi))*(np.exp(-((x-m)**2)/(2*(s**2))))
plt.plot(x,y,label='μ=0,σ=1')
######
m=0
s=1.5
y = 1/(s*np.sqrt(2*np.pi))*(np.exp(-((x-m)**2)/(2*(s**2))))
plt.plot(x,y,label='μ=0,σ=1.5')
######
m=0
s=2
y = 1/(s*np.sqrt(2*np.pi))*(np.exp(-((x-m)**2)/(2*(s**2))))
plt.plot(x,y,label='μ=0,σ=2')

plt.text(-10,1,r'$\mathtt{\phi(x,\mu,\sigma) = \frac{1}{\sigma \sqrt{2}}  exp(-\frac{(x-\mu)^2}{2\sigma^2})}$')
plt.legend(loc='upper right')
plt.show()
```
